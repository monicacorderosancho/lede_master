```python
response = requests.get('https://www.applesfromny.com/varieties/')
doc = BeautifulSoup(response.text, 'html.parser')
```

```python
# doc.find_all('div', class_='apple-variety')
# doc.find_all('div', { 'class': 'apple-variety' })
varieties = doc.select(".apple-variety")

rows = []
for variety in varieties:
    print("-----")
    row = {}
    # print(variety)
    row['title'] = variety.select_one("h2").text
    row['sweetness'] = variety.select_one(".scale-icon").get('class')[-1]
    row['desc'] = variety.select_one("p").text
    row['url'] = variety.select_one("a").get('href')
    row['img_url'] = variety.select_one("img").get('src')
    rows.append(row)
rows
```

```python
import pandas as pd

apples = pd.DataFrame(rows)
apples
```

```python
# We are going to give scrape_apple a url
# it is going to give us back a dictionary

# Cut and paste the code we wrote for ONE page and change THREE things
# 1. Indent it all
# 2. Change the url we're scraping
# 3. Change the last line to 'return pd.Series(row)' to send the row back

# Oops had to make all the titles lowercase
def scrape_apple(url):
    response = requests.get(url)
    doc = BeautifulSoup(response.text, 'html.parser')

    titles = doc.select('.entry-content h2')
    row = {}
    for title in titles:
        # row['Profile'] = ....
        # row['Best Uses'] = ....
        row[title.text.lower()] = title.next_sibling.text
    return pd.Series(row)
```

```python
scraped_df = apples.url.apply(scrape_apple)
scraped_df
```
