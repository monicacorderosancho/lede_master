```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
### 1. go and grab this web and .get and all the info I need to know is contained in this url 
### store the information from the URL in doc (a way to named)
response = requests.get("https://www.bbc.com/")
doc = BeautifulSoup(response.text,'html.parser')
```

### 3 ways to call link and headline

```python
## every single link that exist in the pag
## doc.find_all('a')
## use class_
doc.find_all('h3', class_='media__title')
```

```python
doc.select('.media__title')
```

```python
### .media__title means something with the clase media...
### h3 means something with the tag nameoh h3
### 'h3.media__title' means something with the tag h3 and the class clase media...
### wich means we can do crazy things like:
### h3.media__title
### means "a link inside of (an h3 tag with the class of media__title)"
doc.select('h3.media__title')
```

```python
### Get everything with the class of media__title
## and then loop through them each and print out the text without spaces
titles = doc.select('.media__title')
for title in titles:
    print(title.text.strip())
```

```python
tags = doc.select('.media__tag')
for tag in tags:
    print(tag.text.strip())
```

```python
### create columns 
stories = doc.select('.media-list__item')
rows = []

for story in stories: 
    print('----')
    
    row = {}
    
    row['title'] = story.select_one('h3').text.strip()
    
    try: 
        row['href'] = story.select_one('.media__link, .real__link')['href']
   
    except:
        print("Couldn't find a link")
    
    try:
        row['tag'] = story.select_one('.media__tag').text.strip()
    except:
        print("Couldn't find a tag!")
    
    try:
        row['summary'] = story.select_one('.media__summary').text.strip()
    except:
        print("Couldn't find a summary!")
       
    print(row)
    
    rows.append(row)
        
```

```python
df = pd.DataFrame(rows)
df

```
