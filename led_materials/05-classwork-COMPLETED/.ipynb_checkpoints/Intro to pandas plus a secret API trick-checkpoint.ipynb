```python
# I want to know total GDP for the country
df.gdp_per_capita * df.population
```

```python
# Let's save it as a new column
# For a new column YOU HAVE TO DO df['gdp'] with quotes
df['gdp'] = df.gdp_per_capita * df.population
df.head()
```

```python
# Compare each country's continent to Asia
# aka "Are you a country with the continent Asia?"
df.continent == 'Asia'
```

```python
# Here are all of the countries in asia
# instead of an if statement, you put
# a comparison inside of df[ .... ]
# It's a filter
df[df.continent == 'Asia']
```

```python
# for AND you use parens and &
# for OR you use parens and |
df[(df.continent == 'Asia') & (df.population > 1000000000)]
```
