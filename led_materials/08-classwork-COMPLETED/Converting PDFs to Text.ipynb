````python
from pdfminer.high_level import extract_text
````

````python
text = extract_text('new-york/nys-bill.pdf')
````

````python
text[:1000]
````

````python
import glob
````

````python
filenames = glob.glob("new-york/nys-bill*.pdf")
filenames
````

````python
contents = [extract_text(filename) for filename in filenames]
len(contents)
````

````python
import pandas as pd

pd.DataFrame({
    'filename': filenames,
    'contents': contents
})
````
