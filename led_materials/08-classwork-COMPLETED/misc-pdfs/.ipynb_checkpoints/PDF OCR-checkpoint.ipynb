# Apache Tika

```python
import tika
tika.initVM()

from tika import parser
parsed = parser.from_file('__players-scan-version-1.pdf')
```

```python
parsed
```

```python
from pdf2image import convert_from_path
pages = convert_from_path('__players-scan-version-1.pdf', 300)
for page in pages:
    page.save('file.tiff')
    parsed = parser.from_file('file.tiff', headers={
        'extractInlineImages': 'true'
    })
```

```python
parsed
```

```python
from pdf2image import convert_from_path
pages = convert_from_path('__players-scan-version-1.pdf', 300)
for page in pages:
    page.save('file.tiff')
    parsed = parser.from_file('file.tiff')
```

```python
parsed
```
